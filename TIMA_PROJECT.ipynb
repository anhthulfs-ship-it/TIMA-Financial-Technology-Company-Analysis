```python
import pandas as pd
csv_file_path = '/content/drive/MyDrive/Database for DA project/Tima_CRM - Sheet1.csv'
df = pd.read_csv(csv_file_path)
# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
# Check the number of missing values in each column
print("Number of missing values by column:")
print(df.isnull().sum())
```

#Remove Duplicate Rows

```python
print(f"DataFrame shape before dropping duplicates: {df.shape}")

df = df.drop_duplicates()

print(f"DataFrame shape after dropping duplicates: {df.shape}")
display(df.head())
```

#Delete 4 first columns which are not neccessary and duplicated

```python
# Drop 4 unnecessary columns
df = df.drop(columns=['STT', 'SoTienDKVayBanDau', 'TienGiaiNgan', 'SoTienConLai'])

print("✅ 4 columns have been removed.")
print("New shape:", df.shape)
```

#Current Data Types of Columns

```python
df.info()
```

#Converting Date Columns to Datetime Type

```python
date_columns = ['application_date', 'FromDate', 'ToDate', 'Birthday', 'CheckTime']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print("Data types after conversion:")
df.info()
```

#Converting Numerical Columns to Appropriate Types

```python
float_cols = ['Số tiền đăng ký vay ban đầu', 'Tiền giải ngân', 'Tiền gốc còn lại']
for col in float_cols:
    df[col] = df[col].astype('float64')

# Convert 'Salary' to float64, coercing errors to NaN
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

print("Data types after numerical conversion:")
df.info()
```

#Converting Object Columns to Category Type

```python
category_cols = ['Thời gian đã sống', 'Gender', 'Trạng thái', 'ProductCreditName', 'InterestPaymentType', 'Hình thức cư trú', 'ReceiveYourIncomeSalary', 'HasBadDebt', 'HasLatePayment']

for col in category_cols:
    df[col] = df[col].astype('category')

print("Data types after category conversion:")
df.info()
```

#Converting Remaining Object Columns to String Type

```python
string_cols = ['ID', 'LoanID', 'Số điện thoại khách hàng', 'FullName', 'CityName', 'DistrictName', 'WardName', 'Street', 'CityNameHouseHold', 'DistrictNameHouseHold', 'WardNameHouseHold', 'JobName', 'NameCompany', 'AddressCompany', 'CityCompany', 'DistrictNameCompany', 'DescriptionPositionJob', 'RelativeFamilyName', 'FullNameFamily', 'CreditInfo', 'Name', 'Address', 'Brieft']

df[string_cols] = df[string_cols].astype('string')

print("Covert to string dtype")

print(df[string_cols].dtypes)
```

#Processing GENDER, HASBADDEBT, HASLATEPAYMENT column

```python
# ====================== PROCESSING GENDER COLUMN ======================

# Step 1: Convert to string and remove extra spaces
df['Gender'] = df['Gender'].astype(str).str.strip()

# Step 2: Replace values and handle NaN
df['Gender'] = df['Gender'].replace({
    '0': 'Male',
    '1': 'Female',
    'nan': 'Unknown',
    'NaN': 'Unknown',
    'None': 'Unknown'
})

# Step 3: Fill any remaining NaN values
df['Gender'] = df['Gender'].fillna('Unknown')

# Step 4: Convert to category type (best practice for categorical data)
df['Gender'] = df['Gender'].astype('category')

# ====================== CHECK RESULT ======================
print("✅ Gender column after processing:")
print(df['Gender'].value_counts(dropna=False))
print("\nData type:", df['Gender'].dtype)
```

```python
# ====================== PROCESSING HasBadDebt & HasLatePayment ======================

# Processing HasBadDebt
df['HasBadDebt'] = df['HasBadDebt'].astype(str).str.strip().replace({
    '0': 'No',
    '1': 'Yes'
}).astype('category')

# Processing HasLatePayment
df['HasLatePayment'] = df['HasLatePayment'].astype(str).str.strip().replace({
    '0': 'No',
    '1': 'Yes'
}).astype('category')

# ====================== Check result ======================
print("✅ Kết quả sau khi xử lý 2 cột:\n")

print("HasBadDebt value counts:")
print(df['HasBadDebt'].value_counts(dropna=False))

print("\nHasLatePayment value counts:")
print(df['HasLatePayment'].value_counts(dropna=False))

print("\nData types:")
print(df[['HasBadDebt', 'HasLatePayment']].dtypes)
```

#Standardize Text Columns

```python
# List of columns to standardize
text_cols = ['FullName', 'Name', 'RelativeFamilyName', 'FullNameFamily',
             'NameCompany', 'Street', 'Address', 'AddressCompany',
             'DescriptionPositionJob', 'JobName', 'Brief']

# Apply standardization
for col in text_cols:
    if col in df.columns:
        # Convert to string and remove extra spaces
        df[col] = df[col].astype(str).str.strip()

        # Replace real NaN values with 'Unknown'
        df[col] = df[col].replace(['nan', 'NaN', 'None', 'NAN'], 'Unknown')

        # Fill any remaining NaN
        df[col] = df[col].fillna('Unknown')

        # Convert to Title Case
        df[col] = df[col].str.title()
        print(f"✓ Đã chuẩn hóa cột: {col}")
    else:
        print(f"⚠ Cột {col} không tồn tại")

# Check result
print("\n===SAMPLE RESULTS AFTER PROCESSING===")
for col in ['FullName', 'NameCompany', 'Street']:
    if col in df.columns:
        print(f"\n{col}:")
        print(df[col].head(10))
```

#Fill missing values and handle data

```python
# List of columns to fill NaN with 'Unknown'
fill_unknown_cols = [
    'WardName',
    'Hình thức cư trú',
    'Thời gian đã sống',
    'CityCompany',
    'DistrictNameCompany',
    'ReceiveYourIncomeSalary',
    'CreditInfo',
    'CheckTime'
]

# Fill missing values
for col in fill_unknown_cols:
    if col in df.columns:
        missing_count = df[col].isnull().sum()

        df[col] = df[col].fillna('Unknown')

        print(f"✓ Filled {missing_count} missing values in '{col}' with 'Unknown'")
    else:
        print(f"⚠ Column '{col}' not found in DataFrame")

# ====================== CHECK RESULT ======================
print("\nCompleted filling missing values with 'Unknown'")

# Summary of remaining missing values
print("\nRemaining missing values in selected columns:")
print(df[fill_unknown_cols].isnull().sum())
```

```python
# ====================== FILL MISSING VALUES ======================

# List of columns to fill with 'Unknown'
fill_unknown_cols = [
    'HouseholdDistrict',
    'HouseholdCity',
    'HouseholdWard',
    'Birthday',
    'application_date',
    'CheckTime'
]

# Fill 'Unknown' for selected columns
for col in fill_unknown_cols:
    if col in df.columns:
        missing_count = df[col].isnull().sum()
        df[col] = df[col].fillna('Unknown')
        print(f"✓ Filled {missing_count} missing values in '{col}' with 'Unknown'")

# Special handling for ProductCreditName - Fill with mode (most frequent value)
if 'ProductCreditName' in df.columns:
    mode_value = df['ProductCreditName'].mode()[0]  # Get the most common value
    missing_count = df['ProductCreditName'].isnull().sum()

    df['ProductCreditName'] = df['ProductCreditName'].fillna(mode_value)
    print(f"✓ Filled {missing_count} missing values in 'ProductCreditName' with mode: '{mode_value}'")

# ====================== FINAL CHECK ======================
print("\n" + "="*70)
print("MISSING VALUES FILLING COMPLETED!")

print("\nRemaining missing values:")
print(df[['HouseholdDistrict', 'HouseholdCity', 'HouseholdWard',
          'Birthday', 'application_date', 'ProductCreditName']].isnull().sum())
```

```python
#CLEAN & FILL SALARY COLUMN

# Step 1: Convert Salary to string first
df['Salary'] = df['Salary'].astype(str).str.strip()

# Step 2: Clean the column (remove special characters, $ , commas, etc.)
df['Salary'] = df['Salary'].replace({
    r'[$,]': '',           # Remove $ and comma
    r'Ten thousand': '10000',
    r'thousand': '000',
    r'[a-zA-Z]': '',       # Remove all letters
}, regex=True)

# Step 3: Convert to numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Step 4: Fill missing values with median
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

print("Đã làm sạch và điền Salary thành công!")
print("Median Salary:", median_salary)
print("Remaining missing values:", df['Salary'].isnull().sum())
print("\nSummary:")
print(df['Salary'].describe().round(2))
```

```python
#HANDLE SALARY = 0

print("Số lượng Salary = 0 trước khi xử lý:", (df['Salary'] == 0).sum())

# Tính median của những người có lương > 0
median_salary = df[df['Salary'] > 0]['Salary'].median()

# Thay Salary = 0 bằng median
df['Salary'] = df['Salary'].replace(0, median_salary)

print(f"Đã thay { (df['Salary'] == 0).sum() } giá trị Salary = 0 bằng median: {median_salary:,.0f}")
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in numeric_cols:
    unique_vals = sorted(df[col].dropna().unique())
    unique_count = len(unique_vals)

    print(f"🔹 Cột: {col}  |  Unique: {unique_count}")
    print(f"   Min: {df[col].min():,.2f}  |  Max: {df[col].max():,.2f}")

    if unique_count <= 50:
        print("   All unique values:", unique_vals)
    else:
        print("   First 15 values :", unique_vals[:15])
        print("   Last 15 values  :", unique_vals[-15:])

    print("-" * 80)
```

```python
print("\nNumberOfLoans - Giá trị âm:")
print(df[df['NumberOfLoans'] < 0]['NumberOfLoans'].value_counts())
```

```python
#HANDLE NEGATIVE VALUES IN NUMERICAL COLUMNS

# Check negative values before processing
print("BEFORE PROCESSING:")
print(f"NumberOfLoans has negative values: {(df['NumberOfLoans'] < 0).sum()} records")

# Process only NumberOfLoans: Convert negative values to 0
# (Number of loans cannot be negative)
df['NumberOfLoans'] = df['NumberOfLoans'].clip(lower=0)

print("\n PROCESSING COMPLETED!")
print(f"NumberOfLoans has negative values remaining: {(df['NumberOfLoans'] < 0).sum()}")

#SUMMARY AFTER PROCESSING
print("\n=== SUMMARY AFTER PROCESSING ===")
print(df['NumberOfLoans'].describe().round(2))
```

#Remane columns to English

```python
#RENAME COLUMNS TO ENGLISH

# Define mapping from Vietnamese to English column names
column_mapping = {
    'Số tiền đăng ký vay ban đầu': 'InitialLoanAmount',
    'Tiền giải ngân': 'DisbursedAmount',
    'Tiền gốc còn lại': 'RemainingPrincipal',
    'Số điện thoại khách hàng': 'PhoneNumber',
    'Hình thức cư trú': 'ResidenceType',
    'Thời gian đã sống': 'YearsOfResidence',
    'NameCompany': 'CompanyName',
    'AddressCompany': 'CompanyAddress',
    'CityCompany': 'CompanyCity',
    'DistrictNameCompany': 'CompanyDistrict',
    'RelativeFamilyName': 'RelativeFamilyName',
    'FullNameFamily': 'FamilyFullName',
    'InterestPaymentType': 'InterestPaymentType',
    'Trạng thái': 'Status',
    'CityNameHouseHold': 'HouseholdCity',
    'DistrictNameHouseHold': 'HouseholdDistrict',
    'WardNameHouseHold': 'HouseholdWard'
}

# Apply the new column names
df = df.rename(columns=column_mapping)

print("Successfully renamed all columns to English!")
print("\nNew column names:")
print(df.columns.tolist())
```

```python
#COMPREHENSIVE COLUMN CHECK

print("📋 COMPREHENSIVE DATAFRAME SUMMARY\n")
print(f"Total rows: {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}\n")

# Create summary table
summary = pd.DataFrame({
    'Column Name': df.columns,
    'Data Type': df.dtypes,
    'Non-Null Count': df.count().values,
    'Missing Count': df.isnull().sum().values,
    'Missing %': (df.isnull().mean() * 100).round(2).values,
    'Unique Values': df.nunique().values
})

# Sort by missing percentage (highest first)
summary = summary.sort_values(by='Missing %', ascending=False).reset_index(drop=True)

# Display the full summary
print(summary.to_string(index=False))

# Additional useful information
print("\n" + "="*80)
print("DATA TYPES SUMMARY:")
print(df.dtypes.value_counts())

print("\n" + "="*80)
print("SAMPLE DATA (first 5 rows):")
display(df.head())
```

```python
# ====================== CLEAN PHONENUMBER & FULLNAME ======================

# Clean PhoneNumber: Only numbers allowed
df['PhoneNumber'] = df['PhoneNumber'].astype(str).str.strip()
df['PhoneNumber'] = df['PhoneNumber'].apply(
    lambda x: x if x.replace(' ', '').replace('+', '').replace('-', '').isdigit() else 'Unknown'
)

# Clean FullName: Only letters allowed
df['FullName'] = df['FullName'].astype(str).str.strip()
df['FullName'] = df['FullName'].apply(
    lambda x: x if not any(char.isdigit() for char in x) else 'Unknown'
)

# ====================== RESULT ======================
print("Cleaning completed!")
print("PhoneNumber unique values:", df['PhoneNumber'].nunique())
print("FullName unique values :", df['FullName'].nunique())
```
